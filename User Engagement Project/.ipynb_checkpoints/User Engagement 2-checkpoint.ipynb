# Brandazon User Engagement Analytics

## Background
Brandazon has recently invested in a major cross-platform strategy to boost engagement and retention across mobile and web. This includes a high-profile 2024 partnership with **Labubu** — a collectible product line driving cultural and purchasing momentum — as well as robust **A/B testing**, **push notification campaigns**, and **product scarcity mechanics**.

Recently, Brandazon launched major initiatives – including an influencer-driven Labubu partnership campaign, extensive A/B testing of features, and targeted promotions – to drive app installs, repeat purchases, and overall user activity.

## Business Objective
Leverage large-scale, multi-channel user interaction data (2023–2025) to:

- Detect behavioral signals that predict **retention**, **conversion**, or **drop-off**
- Evaluate the impact of major initiatives:
  - The **Labubu partnership campaign**
  - Seasonal and holiday promotional surges
  - **A/B tests** around scarcity, product placement, and messaging
- Compare mobile and web engagement dynamics
- Surface actionable insights to:
  - Improve **onboarding**
  - Increase **purchase frequency**
  - Optimize **campaign targeting**
  - Maximize **lifetime value**

## Stakeholders

- **Marketing and Growth Teams**  
  Understand which campaigns and creatives drive clicks, opens, and conversions.

- **Product Managers and UX Designers**  
  Use behavioral funnels and A/B test results to refine app and web features.

- **Customer Success and Retention Teams**  
  Monitor repeat purchases and build interventions for at-risk users.

- **Executive Leadership**  
  Measure the ROI of large initiatives like the Labubu partnership.

- **End Users and Customers**  
  Benefit from a more personalized, intuitive, and responsive Brandazon experience.

```python
import pandas as pd
import numpy as np
import os, re
from pandas import json_normalize

# --- Configuration ---
csv_files = [
    "brandazon_identify_preview.csv",
    "brandazon_page_preview.csv",
    "brandazon_screen_preview.csv",
    "brandazon_track_preview.csv"
]

# --- Optional unmatched files (if you want to include these later) ---
unmatched_files = []
all_dataframes = []

# --- Helper: Categorize files by year and type ---
def organize_files_by_year_and_type(file_list):
    files_by_year = {}
    for filename in file_list:
        match = re.search(r'(track|identify|page|screen).*?(\d{4})?', filename, re.IGNORECASE)
        if match:
            event_type = match.group(1).lower()
            year = int(match.group(2)) if match.group(2) else None
            files_by_year.setdefault(year, {}).setdefault(event_type, []).append(filename)
    return files_by_year

# --- Helper: Load a single file and annotate ---
def load_csv_with_metadata(file_path, year=None, event_type=None):
    try:
        df = pd.read_csv(file_path)  # Removed skiprows=1
        df['source_file'] = file_path
        df['year'] = year
        df['event_type'] = event_type
        return df
    except Exception as e:
        print(f"❌ Error loading {file_path}: {e}")
        return None

# --- Load all structured data ---
def load_event_logs(files):
    files_by_year = organize_files_by_year_and_type(files)
    dataframes = []

    for year, types in files_by_year.items():
        for event_type, file_list in types.items():
            for file_path in file_list:
                df = load_csv_with_metadata(file_path, year, event_type)
                if df is not None:
                    dataframes.append(df)
    return pd.concat(dataframes, ignore_index=True)

# --- Main Load ---
combined = load_event_logs(csv_files)

# --- Timestamp parsing and session mapping ---
combined['timestamp'] = pd.to_datetime(combined['timestamp'], errors='coerce')
combined['session_user'] = combined['userId'].fillna(combined['anonymousId'])

# --- Summary Stats ---
print(f"✅ Combined dataset shape: {combined.shape}")
print("📊 Columns:", combined.columns.tolist())
print(f"🕒 Time span: {combined['timestamp'].min()} → {combined['timestamp'].max()}")
print(f"👤 Unique logged-in users: {combined['userId'].nunique()} | 🕶️ Unique anonymous users: {combined['anonymousId'].nunique()}")
print(f"📈 Total events: {len(combined):,}")
```

```python
# Helper function to parse JSON-like entries robustly
def robust_parse_json(val):
    if isinstance(val, dict):
        return val  # already parsed
    if val is None or (isinstance(val, float) and np.isnan(val)):
        return None
    if isinstance(val, str):
        v = val.strip()
        if v.lower() in ('', 'none', 'null', 'nan'):
            return None
        try:
            return json.loads(v)
        except Exception:
            # As fallback, try a direct eval (not safe in production, but here just to handle single quotes or minor issues)
            try:
                return eval(v)
            except Exception:
                return None
    return None

# Parse 'properties' and 'context' JSON fields if present
if 'properties' in combined.columns:
    combined['properties'] = combined['properties'].apply(robust_parse_json)
if 'context' in combined.columns:
    combined['context'] = combined['context'].apply(robust_parse_json)
```

```python
# Sort by user and time, then compute session breaks (30 min threshold)
combined = combined.sort_values(['session_user', 'timestamp'])
combined['time_diff'] = combined.groupby('session_user')['timestamp'].diff().dt.total_seconds()
combined['new_session'] = (combined['time_diff'].isna()) | (combined['time_diff'] > 1800)
combined['session_id'] = combined.groupby('session_user')['new_session'].cumsum()
```

```python
# Define churn cutoff date (e.g., April 1, 2025)
cutoff_date = pd.to_datetime("2025-04-01")
last_event_by_user = combined.groupby('userId')['timestamp'].max()
# Mark churned = True if last event < cutoff (i.e., user became inactive before Q2 2025)
combined['churned'] = combined['userId'].map(lambda uid: last_event_by_user.get(uid, pd.Timestamp.min) < cutoff_date)
```

```python
purchase_events = combined[
    (combined['event'].str.lower().str.contains('order completed')) |
    (combined['event'].str.lower().str.contains('purchase'))
]
```

```python
# Aggregate user-level features for modeling
user_features = combined.groupby('userId').agg(
    total_events=('event', 'count'),
    session_count=('session_id', 'nunique'),
    used_mobile=('source_file', lambda files: any("Mobile" in str(f) for f in files))
).reset_index()
# Add churn label to this user features table
user_features['churned'] = user_features['userId'].map(lambda uid: combined.loc[combined['userId'] == uid, 'churned'].iloc[0])
print(user_features.head(5))
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, classification_report

# Prepare your features and target variable
# Make sure all features are numeric and without missing values
X = user_features[['total_events', 'session_count', 'used_mobile']].copy()
X['used_mobile'] = X['used_mobile'].astype(int)
y = user_features['churned'].astype(int)

# Split your data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42
)

# Create a modeling pipeline with scaling and classifier placeholder
pipe = Pipeline([
    ('scale', StandardScaler()),   # Applies standard scaling
    ('clf', LogisticRegression(max_iter=1000))  # Placeholder, will be replaced in grid
])

# Define hyperparameter grids for different models
param_grid = [
    # Logistic Regression
    {
        'clf': [LogisticRegression(max_iter=1000, solver='liblinear')],
        'clf__C': [0.01, 0.1, 1.0]
    },
    # Random Forest
    {
        'clf': [RandomForestClassifier(random_state=42)],
        'clf__n_estimators': [100, 200],
        'clf__max_depth': [5, 10]
    },
    # XGBoost
    {
        'clf': [XGBClassifier(eval_metric='logloss', random_state=42)],
        'clf__n_estimators': [50, 100],
        'clf__max_depth': [3, 5]
    }
]

# Run grid search with cross-validation to find the best model/hyperparameters
grid = GridSearchCV(
    pipe,
    param_grid,
    cv=3,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1  # Set to 1 for progress output; 0 for silent
)

grid.fit(X_train, y_train)

print(f"Best Model: {grid.best_params_} with CV AUC={grid.best_score_:.3f}")
best_model = grid.best_estimator_

# Evaluate the best model on the test set
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:, 1]

print("\nTest Set Performance:")
print(classification_report(y_test, y_pred))
print(f"Test ROC AUC: {roc_auc_score(y_test, y_proba):.3f}")
```

```python
# Simulate A/B test grouping (for demonstration purposes)
user_ids = user_features['userId'].values
np.random.seed(42)
ab_map = {uid: np.random.choice(['A','B']) for uid in user_ids}
user_features['ab_variation'] = user_features['userId'].map(ab_map)

# Compare churn rate between A and B groups
ab_churn = user_features.groupby('ab_variation')['churned'].mean()
print("Churn rate by A/B Test Variation:")
print(ab_churn)
ab_churn.plot(kind='bar', color='skyblue', title='Churn Rate by A/B Variation')
plt.ylabel('Churn Rate'); plt.xlabel('Variation Group'); plt.tight_layout(); plt.show()
```

```python
# Flag a random 20% of users as 'Labubu campaign engaged'
np.random.seed(42)
engaged_users = np.random.choice(user_ids, size=int(0.2*len(user_ids)), replace=False)
user_features['labubu_engaged'] = user_features['userId'].isin(engaged_users)

# Compute churn rates for engaged vs not engaged
churn_rates = user_features.groupby('labubu_engaged')['churned'].mean()
print("Churn rate (Labubu Engaged vs Not Engaged):")
print(churn_rates)  # e.g., False: 0.602, True: 0.601 (just an example)
# Statistical significance test (t-test)
from scipy.stats import ttest_ind
churn_engaged = user_features[user_features['labubu_engaged']==True]['churned']
churn_not = user_features[user_features['labubu_engaged']==False]['churned']
stat, p_val = ttest_ind(churn_engaged, churn_not, equal_var=False)
print(f"T-test p-value: {p_val:.5f}")
# Plot churn rates
churn_rates.index = ['Not Engaged','Labubu Engaged']
churn_rates.plot(kind='bar', color=['gray','orange'], title='Churn Rate by Labubu Campaign Engagement')
plt.ylabel('Churn Rate'); plt.xlabel('User Group'); plt.tight_layout(); plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure timestamp is datetime
combined['timestamp'] = pd.to_datetime(combined['timestamp'])

# Get each user's first active day
user_first_seen = combined.groupby('userId')['timestamp'].min().reset_index()
user_first_seen.columns = ['userId', 'first_seen_day']
user_first_seen['first_seen_day'] = user_first_seen['first_seen_day'].dt.floor('D')

# Merge first_seen_day into combined
combined = combined.merge(user_first_seen, on='userId', how='left')

# Calculate event day and days since first activity
combined['event_day'] = combined['timestamp'].dt.floor('D')
combined['days_since_first'] = (combined['event_day'] - combined['first_seen_day']).dt.days

# Drop duplicates: 1 row per user per day
retention_raw = combined[['userId', 'days_since_first']].drop_duplicates()

# Count retained users per day
retention = retention_raw.groupby('days_since_first')['userId'].nunique().reset_index()
retention.columns = ['days_since_first', 'retained_users']

# Get Day 0 user count
day_0 = retention[retention['days_since_first'] == 0]
if not day_0.empty:
    day_0_users = day_0['retained_users'].values[0]
    retention['retention_rate'] = retention['retained_users'] / day_0_users
else:
    print("Warning: No users found on Day 0.")
    retention['retention_rate'] = 0

# Plot retention curve (linear)
plt.figure(figsize=(8,4))
plt.plot(retention['days_since_first'], retention['retention_rate'], marker='o')
plt.xlim(-1, 30)  # Only plot days 0–30
plt.title('User Retention Curve (First 30 Days)')
plt.xlabel('Days Since First Seen')
plt.ylabel('Retention Rate')
plt.grid(True)
plt.tight_layout()
plt.show()

# Print selected retention rates
for day in [1, 7, 30]:
    rate = retention.loc[retention['days_since_first'] == day, 'retention_rate']
    print(f"Day {day} retention rate:", rate.values[0] if not rate.empty else "N/A")

# Plot retention curve (log scale)
plt.figure(figsize=(8,4))
plt.plot(retention['days_since_first'], retention['retention_rate'], marker='o')
plt.yscale('log')
plt.title('User Retention Curve (Log Scale)')
plt.xlabel('Days Since First Seen')
plt.ylabel('Retention Rate (log)')
plt.grid(True, which='both')
plt.tight_layout()
plt.show()
```
