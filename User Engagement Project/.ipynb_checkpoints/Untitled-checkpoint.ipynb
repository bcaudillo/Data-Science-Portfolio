```python
import pandas as pd
import json
import re
import matplotlib.pyplot as plt
import seaborn as sns
import collections
from pandas import json_normalize

# ========== 1. Define file list ==========
csv_files = [
    "Identify 2022.csv", "identify 2024.csv", "identify 2025.csv", "Identify calls 2023.csv",
    "Page 2022.csv", "page 2024.csv", "page 2025.csv", "Page Calls 2023.csv",
    "Track 2022.csv", "track 2024.csv", "track 2025.csv", "Track Calls 2023.csv", "User Engagement Project/User Engagement Web Events with Outliers.csv"
]

# ========== 2. Group files by year and type ==========
files_by_year = {}
for filename in csv_files:
    match = re.search(r'(track|identify|page).*?(\d{4})', filename, re.IGNORECASE)
    if match:
        file_type, year = match.group(1).lower(), int(match.group(2))
        files_by_year.setdefault(year, {}).setdefault(file_type, []).append(filename)

# ========== 3. Helper: Robust context parser ==========
def robust_parse_properties(val):
    if isinstance(val, dict):
        return val
    if val is None or (isinstance(val, float) and pd.isnull(val)):
        return None
    if isinstance(val, str):
        v = val.strip()
        if v.lower() in ['', 'none', 'null', 'nan']:
            return None
        try:
            return json.loads(v)
        except Exception:
            pass
        try:
            return eval(v)
        except Exception:
            return None
    return None
combined_all_years['properties'] = combined_all_years['properties'].apply(robust_parse_properties)

# Step 2: Filter to only rows with product info
product_rows = combined_all_years['properties'].dropna().apply(lambda x: isinstance(x, dict) and 'product_id' in x).index

# Step 3: Build products_df
products_df = combined_all_years.loc[product_rows, [
    'timestamp', 'userId', 'anonymousId', 'event', 'properties'
]].copy()

for key in ['product_id', 'price', 'quantity']:
    products_df[key] = products_df['properties'].apply(lambda x: x.get(key) if isinstance(x, dict) else None)

print(f"\nâœ… products_df shape: {products_df.shape}")
print(products_df.head())

# ========== 4. Load all CSVs ==========
all_dataframes = []
for year, file_types in files_by_year.items():
    for file_type in ['track', 'identify', 'page']:
        for file_path in file_types.get(file_type, []):
            try:
                df = pd.read_csv(file_path, skiprows=1)
                df['source_file'] = file_path
                df['year'] = year
                df['event_type'] = file_type
                all_dataframes.append(df)
            except Exception as e:
                print(f"âŒ Error loading {file_path}: {e}")

# ========== 5. Combine all years ==========
combined_all_years = pd.concat(all_dataframes, ignore_index=True)
print("âœ… Combined shape:", combined_all_years.shape)
print("ðŸ“¦ Columns:", combined_all_years.columns.tolist())

# ========== 6. Debug: Context BEFORE parsing ==========
def debug_context_values(df, n=10):
    print("\n---- First {} raw context values ----".format(n))
    print(df['context'].head(n).tolist())
    print("---- Value counts (top 10) ----")
    print(df['context'].value_counts(dropna=False).head(10))
    string_examples = df['context'][df['context'].apply(lambda x: isinstance(x, str) and x.strip().startswith('{'))].head(3)
    print("---- Example string contexts ----")
    print(string_examples.tolist())

debug_context_values(combined_all_years, n=20)
print("Context column types (before parsing):", collections.Counter(combined_all_years['context'].apply(lambda x: type(x).__name__)))

# ========== 7. Parse context object ==========
combined_all_years['context'] = combined_all_years['context'].apply(robust_parse_context)
print("Context column types (after parsing):", collections.Counter(combined_all_years['context'].apply(lambda x: type(x).__name__)))

# ========== 8. Check context population ==========
has_context = combined_all_years['context'].apply(lambda x: isinstance(x, dict) and len(x) > 0).sum()
empty_context = combined_all_years['context'].apply(lambda x: isinstance(x, dict) and len(x) == 0).sum()
missing_context = combined_all_years['context'].isnull().sum()
print(f"Rows with non-empty context: {has_context}")
print(f"Rows with empty dict context: {empty_context}")
print(f"Rows with missing/None context: {missing_context}")
print(f"Total rows: {len(combined_all_years)}")

# ========== 9. (Optional) Preview non-empty context dicts ==========
non_empty_contexts = combined_all_years['context'][combined_all_years['context'].apply(lambda x: isinstance(x, dict) and len(x) > 0)]
if not non_empty_contexts.empty:
    print("\nExample non-empty context dicts:")
    for val in non_empty_contexts.head(2):
        print(val)
else:
    print("âŒ No non-empty context dicts found in your data.")

# ========== 10. Standard EDA ==========

# --- Parse timestamp, add session_user ---
combined_all_years['timestamp'] = pd.to_datetime(combined_all_years['timestamp'], errors='coerce')
combined_all_years['session_user'] = combined_all_years['userId'].fillna(combined_all_years['anonymousId'])

# --- Unique user calculation ---
user_ids = set(combined_all_years['userId'].dropna())
anon_only_df = combined_all_years[combined_all_years['userId'].isnull()]
pure_anon_ids = set(anon_only_df['anonymousId'].dropna())
total_unique_users = len(user_ids.union(pure_anon_ids))
print(f"\nKnown userIds (not null): {len(user_ids)}")
print(f"Pure anonymousIds (never mapped to userId): {len(pure_anon_ids)}")
print(f"Total unique users (userId + pure anonymousId): {total_unique_users}")

# --- Top events ---
top_events = combined_all_years['event'].value_counts().head(20)
print("\nðŸ”¹ Top Events:\n", top_events)
top_events.plot(kind='barh', figsize=(10, 6), title='Top 20 Events')
plt.xlabel('Count')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# --- Event volume over time ---
combined_all_years.set_index('timestamp').resample('W').size().plot(figsize=(12, 5), title='Weekly Event Volume')
plt.ylabel('Event Count')
plt.xlabel('Week')
plt.tight_layout()
plt.show()

# --- Event vs hour heatmap ---
combined_all_years['hour'] = combined_all_years['timestamp'].dt.hour
heatmap_data = combined_all_years.groupby(['event', 'hour']).size().unstack(fill_value=0)
sns.heatmap(heatmap_data, cmap="YlGnBu")
plt.title("Event Type vs Hour")
plt.xlabel("Hour")
plt.ylabel("Event")
plt.tight_layout()
plt.show()

# --- Session simulation ---
combined_all_years = combined_all_years.sort_values(by=['session_user', 'timestamp'])
combined_all_years['time_diff'] = combined_all_years.groupby('session_user')['timestamp'].diff().dt.total_seconds()
combined_all_years['new_session'] = (combined_all_years['time_diff'] > 1800) | combined_all_years['time_diff'].isnull()
combined_all_years['session_id'] = combined_all_years.groupby('session_user')['new_session'].cumsum()

session_df = combined_all_years.groupby(['session_user', 'session_id']).agg(
    session_start=('timestamp', 'min'),
    session_end=('timestamp', 'max'),
    events=('event', 'count')
).reset_index()
session_df['duration'] = (session_df['session_end'] - session_df['session_start']).dt.total_seconds()
print("ðŸ”¹ Avg events/session:", round(session_df['events'].mean(), 2))
print("ðŸ”¹ Avg session duration (sec):", round(session_df['duration'].mean(), 2), "\n")

# --- User behavior ---
combined_all_years['date'] = combined_all_years['timestamp'].dt.date
unique_days = combined_all_years.groupby('session_user')['date'].nunique()
print("ðŸ”¹ Avg unique active days per user:", round(unique_days.mean(), 2))

user_event_counts = combined_all_years['session_user'].value_counts()
top_20p_cutoff = int(len(user_event_counts) * 0.2)
top_20p_sum = user_event_counts.head(top_20p_cutoff).sum()
total_events = user_event_counts.sum()
print("ðŸ”¹ Top 20% users generate:", round((top_20p_sum / total_events) * 100, 2), "% of events\n")

# --- Time-based trends ---
dow_counts = combined_all_years['timestamp'].dt.day_name().value_counts().reindex([
    "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"
])
dow_counts.plot(kind='bar', title='Events by Day of Week', figsize=(10, 4))
plt.ylabel('Event Count')
plt.tight_layout()
plt.show()

hour_counts = combined_all_years['hour'].value_counts().sort_index()
hour_counts.plot(kind='bar', title='Events by Hour of Day', figsize=(12, 4))
plt.ylabel('Event Count')
plt.xlabel('Hour')
plt.tight_layout()
plt.show()

# --- Unnest context (if present) ---
context_rows = combined_all_years[
    combined_all_years['context'].apply(lambda x: isinstance(x, dict) and len(x) > 0)
][['userId', 'anonymousId', 'context']].copy()

if not context_rows.empty:
    context_flat = json_normalize(context_rows['context'])
    context_flat['userId'] = context_rows['userId'].values
    context_flat['anonymousId'] = context_rows['anonymousId'].values

    print(f"âœ… Flattened context shape: {context_flat.shape}")
    print("ðŸ”Ž Sample columns:", context_flat.columns.tolist()[:10])
    print(context_flat.head())

    # Example: device.model, os.name, campaign, etc.
    if 'device.model' in context_flat.columns:
        print("Top Device Models:\n", context_flat['device.model'].value_counts())
        context_flat['device.model'].value_counts().head(10).plot(kind='bar')
        plt.title('Top Device Models')
        plt.xlabel('Device Model')
        plt.ylabel('User Count')
        plt.tight_layout()
        plt.show()
else:
    print("âŒ No context data found to flatten.")

# -------------- End --------------

```

```python
import json

product_rows = []

for idx, row in combined_all_years.iterrows():
    props = row.get('properties')
    
    # Try to parse props if it's a string
    if isinstance(props, str):
        # Convert single quotes to double quotes for JSON compliance
        props_json_str = props.replace("'", '"')
        try:
            props = json.loads(props_json_str)
        except Exception:
            continue

    # At this point, props is either a dict or None
    if isinstance(props, dict):
        # Products as a list in 'products'
        if 'products' in props and isinstance(props['products'], list):
            for product in props['products']:
                if isinstance(product, str):
                    # Same fix: convert single quotes to double quotes
                    product_json_str = product.replace("'", '"')
                    try:
                        product = json.loads(product_json_str)
                    except Exception:
                        continue
                if isinstance(product, dict):
                    rec = dict(product)
                    rec['year'] = row.get('year')
                    rec['timestamp'] = row.get('timestamp')
                    rec['userId'] = row.get('userId')
                    rec['anonymousId'] = row.get('anonymousId')
                    product_rows.append(rec)
        # Single product as dict at top level
        elif 'product_id' in props:
            rec = dict(props)
            rec['year'] = row.get('year')
            rec['timestamp'] = row.get('timestamp')
            rec['userId'] = row.get('userId')
            rec['anonymousId'] = row.get('anonymousId')
            product_rows.append(rec)

products_df = pd.DataFrame(product_rows)
print("âœ… products_df shape:", products_df.shape)
print(products_df.head())
```

```python
# ---- 6. Filter purchases (adjust event name if needed) ----
# Use the correct event name that represents a purchase in your data!
purchase_event_name = 'Order Completed'
purchases = combined_all_years[combined_all_years['event'] == purchase_event_name].copy()

# ---- 7. Extract campaign info from context ----
purchases.loc[:, 'campaign_source'] = purchases['context'].apply(
    lambda x: x.get('campaign', {}).get('source') if isinstance(x, dict) else None
)
purchases.loc[:, 'campaign_medium'] = purchases['context'].apply(
    lambda x: x.get('campaign', {}).get('medium') if isinstance(x, dict) else None
)
purchases.loc[:, 'campaign_name'] = purchases['context'].apply(
    lambda x: x.get('campaign', {}).get('name') if isinstance(x, dict) else None
)

# ---- 8. Aggregate performance ----
campaign_counts = purchases['campaign_source'].value_counts(dropna=False)
print("\nPurchase counts by campaign source:")
print(campaign_counts)

# ---- 9. Visualize ----
plt.figure(figsize=(10, 5))
campaign_counts.plot(kind='bar', color='orange')
plt.title('Purchases by Campaign Source')
plt.xlabel('Campaign Source')
plt.ylabel('Number of Purchases')
plt.tight_layout()
plt.show()

# ---- 10. Advanced: Purchases by campaign (source + medium) ----
if not purchases.empty:
    campaign_combo = purchases.groupby(['campaign_source', 'campaign_medium']).size().unstack(fill_value=0)
    print("\nPurchases by source + medium:\n", campaign_combo)

    campaign_combo.plot(kind='bar', stacked=True, figsize=(12, 6))
    plt.title('Purchases by Campaign Source + Medium')
    plt.xlabel('Campaign Source')
    plt.ylabel('Number of Purchases')
    plt.tight_layout()
    plt.show()

# ---- 11. Optional: Purchases over time by campaign ----
if 'timestamp' in purchases.columns:
    purchases['timestamp'] = pd.to_datetime(purchases['timestamp'], errors='coerce')
    purchases['month'] = purchases['timestamp'].dt.to_period('M')
    monthly = purchases.groupby(['month', 'campaign_source']).size().unstack(fill_value=0)
    monthly.plot(figsize=(15,6))
    plt.title('Monthly Purchases by Campaign Source')
    plt.xlabel('Month')
    plt.ylabel('Purchases')
    plt.tight_layout()
    plt.show()
```
